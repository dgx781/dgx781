```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_data= TabularDataset(r"C:\Users\DIBYOJIT\OneDrive\Documents\churn_prediction_simple.csv")
train_data.head()
```

```python
label='churn'
train_data[label].describe()
```

```python
predictor= TabularPredictor(label=label).fit(train_data)
```

```python
test_data= r"C:\Users\DIBYOJIT\OneDrive\Documents\churn_test.csv"
y_pred= predictor.predict(test_data)
y_pred.head()
```

```python
predictor.evaluate(train_data)
```
